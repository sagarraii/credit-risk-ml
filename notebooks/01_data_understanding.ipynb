```python
# cols missing valuse more than 40%

missing_pct = df.isnull().mean()*100

# filter
high_missing_cols = missing_pct[missing_pct > 40].sort_values(ascending=False)

# Display the results
print(f"Total columns with >30% missing data: {len(high_missing_cols)}")
print("-" * 30)
print(high_missing_cols)
```

```python
# Identify columns with low missingness (the reliable ones)
reliable_cols = missing_pct[missing_pct <= 40].index.tolist()

print(f"Total Reliable Columns: {len(reliable_cols)}")
print("---")
print(reliable_cols)
```

Conclusion

This dataset consists of 2,260,701 loan records with 151 columns, making it computationally heavy and unsuitable for naive, full-memory processing. As a result, careful column triage and feature planning are essential before any preprocessing or modeling.

An initial missing-value audit revealed that 46 columns contain more than 40% missing data, with several exceeding 90%. These columns are largely associated with hardship programs, settlement information, secondary applicants, and free-text descriptions, indicating that their absence is mostly structural or conditional rather than random. Such columns are either applicable to a small subset of borrowers or represent post-loan events, making them poor candidates for general modeling.

Among the remaining 105 relatively reliable columns, features were classified based on their functional role rather than data type. This analysis identified:

Identifier and internal tracking variables (e.g., id, member_id, url, policy_code) that do not carry predictive value.

A clearly defined target variable (loan_status), which will be isolated from all feature engineering steps.

Multiple data leakageâ€“prone variables (e.g., payment totals, recoveries, outstanding principal) that reflect post-loan outcomes and must be excluded to preserve model validity.

A strong set of core numeric financial features representing borrower income, credit history, and loan characteristics.

Several categorical application features suitable for later encoding.

A subset of high-cardinality textual or location-based columns (e.g., job titles, descriptions, ZIP codes) that require special handling or removal due to sparsity and encoding cost.

Date-based variables were also identified and reserved for future feature extraction rather than direct use.

Overall, this notebook establishes a clear and defensible feature selection strategy, reduces modeling risk due to leakage, and prepares the dataset for efficient downstream processing. The outcome of this analysis is a curated feature universe, which will be formally reduced and cleaned in the next stage (02_data_cleaning.ipynb). This structured approach ensures scalability, interpretability, and alignment with real-world credit risk modeling practices.
